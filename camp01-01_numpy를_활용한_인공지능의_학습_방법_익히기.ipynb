#### 인간 경사하강법 실시

image.png

```python
import numpy as np

x = np.array([ 2,  4,  6,  8])   # 공부시간 : 독립변수
y = np.array([81, 93, 91, 97])   # 시험점수 : 종속변수

a = -5
b = 80
n = 4

y_pred = a * x + b
error = y_pred - y
MSE = np.mean(error ** 2)

a_grad = (2/n) * np.sum(error * x)  # MSE를 a로 편미분한 결과, a지점에서 접선의 기울기

print(f"y = {a}x + {b} 일때")
print(f"접선의 기울기 : {a_grad}")
print(f"오차제곱의 평균(MSE) : {MSE}")
```

#### 경사하강법 자동실시

```python
# b는 80으로 고정하자...
import numpy as np

x = np.array([ 2,  4,  6,  8])   # 공부시간 : 독립변수
y = np.array([81, 93, 91, 97])   # 시험점수 : 종속변수

a = -5
b = 80
n = 4
lr = 0.003    # 학습률

for i in range(1, 21) :
    y_pred = a * x + b
    error = y_pred - y
    MSE = np.mean(error ** 2)

    print(f"{i}번째, {a:.2f}x+{b}, 오차제곱의 합{MSE:.2f}")

    a_grad = (2/n) * np.sum(error * x)

    a = a - lr * a_grad
```

```python
# a, b를 모두 0으로 초기화하고, a, b가 모두 자동 업데이트 되도록...
import numpy as np

x = np.array([ 2,  4,  6,  8])   # 공부시간 : 독립변수
y = np.array([81, 93, 91, 97])   # 시험점수 : 종속변수

a, b = 0, 0

n = 4
lr = 0.003    # 학습률

for i in range(1, 21) :
    y_pred = a * x + b
    error = y_pred - y
    MSE = np.mean(error ** 2)

    print(f"{i}번째, {a:.2f}x+{b:.2f}, 오차 제곱의 합{MSE:.2f}")

    a_grad = (2/n) * np.sum(error * x)
    b_grad = (2/n) * np.sum(error)

    a = a - lr * a_grad
    b = b - lr * b_grad

# a, b 업데이트 하는 코드
# 학습률 조정
```
