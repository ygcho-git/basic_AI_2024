## 람다함수(Lambda Function) : 이름없는 함수

lambda 매개변수 : 리턴값

image.png

```python
# 일반 함수의 정의
def add(a, b) :
    return a + b

def square(x) :
    return x**2

def is_even(x) :
    return x%2 == 0

print(add(1, 2))
print(square(3))
print(is_even(2))
```

```python
# 위와 같은 단순한 함수는 람다함수로 정의하면 간편

lambda a, b : a + b
lambda x : x ** 2
lambda x : x % 2 == 0
```

image.png

```python
# 이 단순한 것을 일반 함수로 정의하면 코드가 길어짐.
def square(x) :
    return x**2

a, b, c = map(square, [1, 2, 3])
print(a, b, c)
```

```python
# 만약 람다함수에 이름을 주고 싶다면, 변수에 저장

square = lambda x : x ** 2

a, b, c = map(square, [1, 2, 3])
print(a, b, c)
```

```python
# map() 같은 고차함수에서 람다함수와 함께 사용하면 간결해짐.

a, b, c = map(lambda x : x ** 2, [1, 2, 3])
print(a, b, c)
```

### 고차원 함수 filter()

#### 짝수만 걸러내기

```python
# 일반함수 사용해서
def is_even(x) :
    return x % 2 == 0

L = [1, 2, 8, 3, 17, 6, 53, 129]

L_even = list(filter(is_even, L))

print(L_even)
```

```python
# 람다함수를 사용해서

L = [1, 2, 8, 3, 17, 6, 53, 129]

L_even = list(filter(lambda x : x % 2 == 0, L))

print(L_even)
```

```python
# 2차원 이상의 데이터를 정렬할 떄는, 정렬기준 함수를 매개변수로 사용

def sort_key(l) :
    return (l[1], -l[0])

L1 = [[1, 1], [2, 4], [2, 2], [3, 2]]
L2 = [[1, 1], [2, 4], [2, 2], [3, 2]]

L1.sort()                               # 기본 정렬기준 : 0번열에 오름차순, 0번 열이 같으면 1번열에 오름차순
L2.sort(key = sort_key)  # 기본 정렬기준 : 1번열에 오름차순, 1번 열이 같으면 0번열에 내림차순

print(*L1)
print(*L2)
```

```python
sort_key = lambda l : (l[1], -l[0])

L1 = [[1, 1], [2, 4], [2, 2], [3, 2]]
L2 = [[1, 1], [2, 4], [2, 2], [3, 2]]

L1.sort()  # 1순위 : 0번열 오름차순, 2순위 : 1번열 오름차순

L2.sort(key = sort_key)
           # 1순위 : 1번열 오름차순, 2순위 : 0번열 내림차순
print(*L1)
print(*L2)
```

#### 은행 거리 클래스 정의

image.png

```python
# 입금, 출금 메서드

class BankAccount:
    def __init__(self, owner, money=0):
        self.owner = owner       # 예금자 이름
        self.money = money       # 잔액

    def deposit(self, amount):              # 입금 메서드
        self.money += amount

    def withdraw(self, amount):             # 출금 메서드
        if self.money >= amount:
            self.money -= amount
        else :
            print("잔액 부족")

seo = BankAccount("서연", 1000)     # 서연이 계좌 객체 생성
hyun = BankAccount("현솔", 800)     # 현솔이 계좌 객체 생성

seo.withdraw(200)                    # 서연이 계좌에서 200원 출금
hyun.deposit(300)                    # 현솔이 계좌에 300원 입금

print(f"{seo.owner}님 계좌{seo.money}원")
print(f"{hyun.owner}님 계좌{hyun.money}원")
```

```python
# 계좌이체 메서드

class BankAccount:
    def __init__(self, owner, money=0):
        self.owner = owner       # 예금자 이름
        self.money = money       # 잔액

    def deposit(self, amount):              # 입금 메서드
        self.money += amount

    def withdraw(self, amount):             # 출금 메서드
        if self.money >= amount:
            self.money -= amount
        else :
            print("잔액 부족")

    def transfer(self, other, amount):      # 계좌 이체 메서드
        if  self.money >= amount:
            self.money -= amount
            other.money += amount
        else:
            print("잔액 부족")


seo = BankAccount("서연", 1000)     # 서연이 계좌 객체 생성
hyun = BankAccount("현솔", 800)     # 현솔이 계좌 객체 생성

seo.transfer(hyun, 100)        # 서연이가 현솔이에게 100원 예금 이체

print(f"{seo.owner}님 계좌{seo.money}원")
print(f"{hyun.owner}님 계좌{hyun.money}원")
```

#### 동물 달리기 클래스 정의(토끼와 거북이의 경주)



image.png

```python
class Animal:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed    # 초속
        self.position = 0

    def run(self):                     # 달리기 메서드(위치 변경)
        self.position += self.speed

    def race(self, other, distance):   # 달리기 경주 메서드
        self.position = 0
        other.position = 0

        seconds = 0
        while self.position < distance  and other.position < distance :
            seconds += 1
            self.run()
            other.run()
            print(f"{seconds}초:",
                  f"{self.name} {self.position}m,",
                  f"{other.name} {other.position}m")

rabbit = Animal("토끼", 10)
turtle = Animal("거북이", 2)
rabbit.race(turtle, 50)                 # 경주 거리는 50m라는 의미
```

#### kNN 알고리즘

**px**, **py**와 가장 가까운 점 3개를 거리가 가장 가까운 것부터 출력?

image.png

```python
D = [[(21, 20), 0],
     [(22, 18), 0],
     [(19, 17), 0],
     [(23, 20), 0],
     [(18, 18), 0],
     [(22, 21), 0],
     [(21, 19), 0]]

px, py = map(int, input().split())

for i in range(7) :
    x, y = D[i][0][0], D[i][0][1]
    D[i][1] = ((px-x)**2 + (py-y)**2)**0.5

D.sort(key=lambda l : l[1])

for i in range(3) :
    print(f"{D[i][0]}, 거리 : {D[i][1]:.2f}")
```

#### 카탈란 수열 - 다이나믹 프로그래밍

카탈란 수열을 다이나믹 프로그래밍으로
image.png

```python
n = int(input())

d = [[0 for _ in range(n+1)] for _ in range(n+1)]

for i in range(n+1):
    d[i][0] = 1

for i in range(0, n+1):
    for j in range(1, i+1):
        d[i][j] = d[i-1][j] + d[i][j-1]

print(f"{d[n][n]}가지")
```
