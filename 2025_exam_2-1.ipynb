### 컨테이너 자료형의 종류 및 기본 특징

```python
# 컨테이너 자료형의 생성
A = [2, 3, 5, 4, 1]
B = (2, 3, 5, 4, 1)
C = "Hello"
D = {1:'H',2:'He',3:'Li'}
E = {2, 3, 5, 4, 1}
```

```python
# 컨테이너 자료형은 모두 in을 통해 원소 존재 유무 확인
print(3 in A)
print(3 in B)
print('o' in C)
print(1 in D)
print('He' in D.values())
print(3 in E)
```

```python
# 컨테이너 자료형은 모두 "반복 가능(iterable) 객체"
for item in A :
    print(item, end=' ')
print()

for item in B :
    print(item, end=' ')
print()

for item in C :
    print(item, end=' ')
print()

for item in D :
    print(item, end=' ')
print()

for item in E :
    print(item, end=' ')
```

```python
# index로 반복하는 것은 "시퀀스 객체"만 가능
for i in range(len(A)) :
    print(A[i], end=' ')
print()

for i in range(len(B)) :
    print(B[i], end=' ')
print()

for i in range(len(C)) :
    print(C[i], end=' ')
print()

# for i in range(len(D)) :
#     print(D[i], end=' ')
# print()

# for i in range(len(E)) :
#     print(E[i], end=' ')
# print()
```

```python
# set()은 원소 중복을 허용하지 않음
# 각 메뉴별 주문 수량?
order = ['치킨', '피자', '치킨', '피자', '피자', '치킨', '떡볶이', '족발', '떡볶이', '치킨']

uniq_order = set(order)

for item in uniq_order :
    print(item, order.count(item))
```

### mutable 객체와 immutable 객체

image.png

```python
# mutable(수정가능) - 해당 메모리 공간을 수정가능하다.
A = [1, 2, 3, 4, 5]
print(A, id(A))

A[1] = 10
print(A, id(A))
```

```python
# immutable(수정 불가능) - 해당 메모리 공간을 수정불가하다.
#                       그래서 수정하려면, 새 메모리 공간에서 작업한다.
A = (1, 2, 3, 4, 5)
print(A, id(A))

# A[0] = 10   # immutable한 객체는 수정이 불가능

A = (10, 2, 3, 4, 5)
print(A, id(A))
```

### 참조 vs 얕은 복사

image.png

```python
A = [1, 2, 3]  # 객체생성
B = A          # 참조
C = A[:]       # 얕은 복사

A[0] = 10
B[1] = 20
C[2] = 30

print(A, B, C)
```

## 람다함수(Lambda Function) : 이름없는 함수

lambda 매개변수 : 리턴값

image.png

```python
# 일반 함수의 정의
def add(a, b) :
    return a + b

def square(x) :
    return x**2

def is_even(x) :
    return x%2 == 0

print(add(1, 2))
print(square(3))
print(is_even(2))
```

```python
# 위와 같은 단순한 함수는 람다함수로 정의하면 간편

lambda a, b : a + b
lambda x : x ** 2
lambda x : x % 2 == 0
```

image.png

```python
# 이 단순한 것을 일반 함수로 정의하면 코드가 길어짐.
def square(x) :
    return x**2

a, b, c = map(square, [1, 2, 3])
print(a, b, c)
```

```python
# 만약 람다함수에 이름을 주고 싶다면, 변수에 저장

square = lambda x : x ** 2

a, b, c = map(square, [1, 2, 3])
print(a, b, c)
```

```python
# map() 같은 고차함수에서 람다함수와 함께 사용하면 간결해짐.

a, b, c = map(lambda x : x ** 2, [1, 2, 3])
print(a, b, c)
```

### 고차원 함수 map()

```python
# 고차원함수 map()
x1, y1, x2, y2 = map(abs, [-5, 2, 5, -2])

print(x1, y1, x2, y2)
```

### 고차원 함수 filter()

#### 짝수만 걸러내기

```python
# 일반함수 사용해서
def is_even(x) :
    return x % 2 == 0

L = [1, 2, 8, 3, 17, 6, 53, 129]

L_even = list(filter(is_even, L))

print(L_even)
```

```python
# 람다함수를 사용해서

L = [1, 2, 8, 3, 17, 6, 53, 129]

L_even = list(filter(lambda x : x % 2 == 0, L))

print(L_even)
```

#### 'a'로 시작하는 과일만 걸러내기

```python
# a로 시작하는 과일은?

def starts_with_a(word) :
    return  word[0] == 'a'

fruit = ["apple", "banana", "avocado", "grape", "apricot"]

a_words = list(filter(starts_with_a, fruit))

print(a_words)
```

```python
# a로 시작하는 과일은?

fruit = ["apple", "banana", "avocado", "grape", "apricot"]

a_words = list(filter(lambda w : w[0] == 'a', fruit))

print(a_words)
```

- sorted() : 내장함수, **모든 컨테이너** 자료형을 정렬할 수 있음. S = sorted({2, 3, 1})

callback함수를 써서 정렬해야 하는 경우

image.png

```python
L = [['대한민국', 13, 9, 10],
     ['미국',     40, 44, 42],
     ['네덜란드', 15, 7, 12],
     ['중국',     40, 27, 24],
     ['이탈리아', 12, 13, 15],
     ['독일',     12, 13, 8],
     ['프랑스',   16, 26, 22],
     ['호주',     18, 19, 16],
     ['일본',     20, 12, 13],
     ['영국',     14, 22, 29]]

L.sort()

for i in range(len(L)) :
    print(L[i])
```

```python
# 2차원 이상의 데이터를 정렬할 떄는, 정렬기준 함수를 매개변수로 사용
L1 = [[1, 1], [2, 4], [2, 2], [3, 2]]
L2 = [[1, 1], [2, 4], [2, 2], [3, 2]]

L1.sort()                               # 기본 정렬기준 : 0번열에 오름차순, 0번 열이 같으면 1번열에 오름차순
L2.sort(key = lambda l: (l[1], -l[0]))  # 기본 정렬기준 : 1번열에 오름차순, 1번 열이 같으면 0번열에 내림차순

print(*L1)
print(*L2)
```

```python
# 2차원 이상의 데이터를 정렬할 떄는, 정렬기준 함수를 매개변수로 사용

def sort_key(l) :
    return (l[1], -l[0])

L1 = [[1, 1], [2, 4], [2, 2], [3, 2]]
L2 = [[1, 1], [2, 4], [2, 2], [3, 2]]

L1.sort()                               # 기본 정렬기준 : 0번열에 오름차순, 0번 열이 같으면 1번열에 오름차순
L2.sort(key = sort_key)  # 기본 정렬기준 : 1번열에 오름차순, 1번 열이 같으면 0번열에 내림차순

print(*L1)
print(*L2)
```

```python
sort_key = lambda l : (l[1], -l[0])

L1 = [[1, 1], [2, 4], [2, 2], [3, 2]]
L2 = [[1, 1], [2, 4], [2, 2], [3, 2]]

L1.sort()  # 1순위 : 0번열 오름차순, 2순위 : 1번열 오름차순

L2.sort(key = sort_key)
           # 1순위 : 1번열 오름차순, 2순위 : 0번열 내림차순
print(*L1)
print(*L2)
```

```python
# sorted도 고차원 함수이다.
T = ((1, 1), (2, 4), (2, 2), (3, 2)) # 튜플은 sort()를 쓸 수 없음...
L = sorted(T, key=lambda t: (t[1], -t[0]))

print(L)
```
