## numpy 배열의 정보 추출

image.png

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]]) #numpy 객체 생성

print(arr) # numpy 배열 출력

print(arr.ndim)     # 2 : numpy 배열의 축의 개수. 이는 배열의 차원이 됨.
print(arr.shape)    # (2, 3) : 배열 각 축의 크기
print(arr.size)     # 6 : 배열의 원소 개수 출력
```

## 7. numpy 배열의 연산

```python
# 파이썬 리스트의 덧셈?
arr1 = [[5, 6],
        [7, 8]]
arr2 = [[1, 2],
        [3, 4]]

print(arr1 + arr2)
```

```python
# numpy 배열의 연산
arr1 = np.array([[5, 6],
                 [7, 8]])
arr2 = np.array([[1, 2],
                 [3, 4]])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
# 행렬의 곱셈
arr1 = np.array([[5, 6],
                 [7, 8]])
arr2 = np.array([[1, 2],
                 [3, 4]])

print(arr1 @ arr2)
```

## 8. Broadcasting(차원이 다른 배열 연산할 때 자동으로 일어남)

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr + 3)
```
