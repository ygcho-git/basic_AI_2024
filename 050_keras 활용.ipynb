```python
# 경사하강법을 통해 a, b를 업데이트 하는 코드
import numpy as np

X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])

a = 0  # a를 0으로 초기화
b = 0  # b를 0으로 초기화
n = len(X)
lr = 0.3    # 학습률
epoch = 2000  # 학습할 횟수

for i in range(1, epoch+1) :
    Y_pred = a * X + b
    error = Y_pred - Y
    MSE = np.mean(error ** 2)

    print(f"{i} epoch - {a:.1f}x + {b:.1f}일 때, >>>>> 오차(error)는 {MSE}")

    a_grad = (2/n) * np.sum(error * X)
    b_grad = (2/n) * np.sum(error)

    a = a - lr * a_grad
    b = b - lr * b_grad
```

## 이제부터 $ax+b$가 아니라, $wx + b$로 하자...
$w$ : weight(가중치)

$b$ : bias(편향)

```python
# 경사하강법을 통해 w, b를 업데이트 하는 코드
import numpy as np

X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])

w = 0  # w를 0으로 초기화
b = 0  # b를 0으로 초기화
n = len(X)
lr = 0.03    # 학습률
epoch = 1000  # 학습할 횟수

for i in range(1, epoch+1) :
    Y_pred = w * X + b
    error = Y_pred - Y
    MSE = np.mean(error ** 2)

    print(f"{i} epoch - {w:.1f}x + {b:.1f}일 때, >>>>> 오차(error)는 {MSE}")

    w_grad = (2/n) * np.sum(error * X)
    b_grad = (2/n) * np.sum(error)

    w = w - lr * w_grad
    b = b - lr * b_grad
```
